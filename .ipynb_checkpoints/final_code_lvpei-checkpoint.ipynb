# Face detection using Cognitive Services
This walkthrough shows you how to use the cognitive services [Face API](https://azure.microsoft.com/services/cognitive-services/face/) to detect faces in an image. The API also returns various attributes such as the gender and age of each person. The sample images used in this walkthrough are from the [How-Old Robot](http://www.how-old.net) that uses the same APIs.

You can run this example as a Jupyter notebook on [MyBinder](https://mybinder.org) by clicking on the launch Binder badge: 

[![Binder](https://mybinder.org/badge.svg)](https://mybinder.org/v2/gh/Microsoft/cognitive-services-notebooks/master?filepath=FaceAPI.ipynb)

For more information, see the [REST API Reference](https://westus.dev.cognitive.microsoft.com/docs/services/563879b61984550e40cbbe8d/operations/563879b61984550f30395236).

## Prerequisites
You must have a [Cognitive Services API account](https://docs.microsoft.com/azure/cognitive-services/cognitive-services-apis-create-account) with **Face API**. The [free trial](https://azure.microsoft.com/try/cognitive-services/?api=face-api) is sufficient for this quickstart. You need the subscription key provided when you activate your free trial, or you may use a paid subscription key from your Azure dashboard.

The next few lines of code call into the Face API to detect the faces in the image. In this instance, the image is specified via a publically visible URL. You can also pass an image directly as part of the request body. For more information, see the [API reference](https://westus.dev.cognitive.microsoft.com/docs/services/563879b61984550e40cbbe8d/operations/563879b61984550f30395236). 

```python
import requests
import math
from IPython.display import HTML

headers = { 'Ocp-Apim-Subscription-Key': subscription_key }
    
params = {
    'returnFaceId': 'true',
    'returnFaceLandmarks': 'true',
    'returnFaceAttributes': 'age,gender,headPose,smile,facialHair,glasses,emotion,hair,makeup,occlusion,accessories,blur,exposure,noise',
}

response = requests.post(face_api_url, params=params, headers=headers, json={"url": image_url})
faces = response.json()
HTML("<font size=5>Detected <font color='blue'>%d</font> faces in the image</font>"%len(faces))
#print(faces[0]['faceLandmarks'])
print("Pupil Left x")
print(faces[0]['faceLandmarks']['pupilLeft']['x'])
print("Pupil Right x")
print(faces[0]['faceLandmarks']['pupilRight']['x'])
print("Eye Left Inner x")
print(faces[0]['faceLandmarks']['eyeLeftInner']['x'])
print("eye Right Inner x")
print(faces[0]['faceLandmarks']['eyeRightInner']['x'])
print("eye Left Bottom x")
ELB_X = faces[0]['faceLandmarks']['eyeLeftBottom']['x']
print(faces[0]['faceLandmarks']['eyeLeftBottom']['x'])

print("eye Left Bottom x")
ERB_X = faces[0]['faceLandmarks']['eyeRightBottom']['x']
print(faces[0]['faceLandmarks']['eyeRightBottom']['x'])
print("eye Left Top x")
ELT_X = faces[0]['faceLandmarks']['eyeLeftTop']['x']
print(faces[0]['faceLandmarks']['eyeLeftTop']['x'])

print("eye Right Top x")
ERT_X = faces[0]['faceLandmarks']['eyeRightTop']['x']
print(faces[0]['faceLandmarks']['eyeRightTop']['x'])
print("eye Left Outer x")
print(faces[0]['faceLandmarks']['eyeLeftOuter']['x'])
print("eye Right Outer x")
print(faces[0]['faceLandmarks']['eyeRightOuter']['x'])

print("Pupil Left y")
print(faces[0]['faceLandmarks']['pupilLeft']['y'])
print("Pupil Right y")
print(faces[0]['faceLandmarks']['pupilRight']['y'])
print("eye Left Inner y")
print(faces[0]['faceLandmarks']['eyeLeftInner']['y'])
print("eye Right Inner y")
print(faces[0]['faceLandmarks']['eyeRightInner']['y'])
print("eye Left Bottom y")
ELB_Y = faces[0]['faceLandmarks']['eyeLeftBottom']['y'];
print(faces[0]['faceLandmarks']['eyeLeftBottom']['y'])
print("eye Right Bottom y")
ERB_Y = faces[0]['faceLandmarks']['eyeRightBottom']['y'];
print(faces[0]['faceLandmarks']['eyeRightBottom']['y'])
print("eye Left Top y")
ELT_Y = faces[0]['faceLandmarks']['eyeLeftTop']['y'];
print(faces[0]['faceLandmarks']['eyeLeftTop']['y'])
print("eye Right Top y")
ERT_Y = faces[0]['faceLandmarks']['eyeRightTop']['y'];
print(faces[0]['faceLandmarks']['eyeRightTop']['y'])
print("eye Left Outer y")
print(faces[0]['faceLandmarks']['eyeLeftOuter']['y'])
print("eye Right Outer y")
print(faces[0]['faceLandmarks']['eyeRightOuter']['y'])

l_temp = ELB_Y-ELT_Y;
r_temp = ERB_Y-ERT_Y;
# print("Left eye")
# print(ELB_Y-ELT_Y)
# print("Right eye")
# print(ERB_Y-ERT_Y)
print("right eye is reference")

l = math.sqrt(math.pow((ELB_X-ELB_X),2) + math.pow((ELT_Y-ELB_Y),2))
r = math.sqrt(math.pow((ERB_X-ERB_X),2) + math.pow((ERT_Y-ERB_Y),2))
# l and r will store fissure height
    
#print(l);
#print(r);
#print("y coordinate difference")
#print(l_temp);
#print(r_temp);
#condition to check find the percentage difference and then assigning it according to Doctor's new grading system
if(l>=r):
    p = (l-r)/l;
    p = p*100;
    #print("%d narrowing",% p);
    print ("%d percentage narrowing" % p)
else:
    p = (r-l)/r;
    p = p*100;
    print("%d percentage narrowing" % p);
    

    
```

Finally, the face information can be overlaid of the original image using the `matplotlib` library in Python.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.lines as mlines


from PIL import Image
from matplotlib import patches
from io import BytesIO

response = requests.get(image_url)
image = Image.open(BytesIO(response.content))

plt.figure(figsize=(40,40))
ax = plt.imshow(image, alpha=0.6)

plt.scatter([faces[0]['faceLandmarks']['pupilLeft']['x']],[faces[0]['faceLandmarks']['pupilLeft']['y']])
plt.scatter([faces[0]['faceLandmarks']['pupilRight']['x']],[faces[0]['faceLandmarks']['pupilRight']['y']])
plt.scatter([faces[0]['faceLandmarks']['eyeLeftTop']['x']],[faces[0]['faceLandmarks']['eyeLeftTop']['y']])
plt.scatter([faces[0]['faceLandmarks']['eyeLeftBottom']['x']],[faces[0]['faceLandmarks']['eyeLeftBottom']['y']])
plt.scatter([faces[0]['faceLandmarks']['eyeRightInner']['x']],[faces[0]['faceLandmarks']['eyeRightInner']['y']])
plt.scatter([faces[0]['faceLandmarks']['eyeRightTop']['x']],[faces[0]['faceLandmarks']['eyeRightTop']['y']])
plt.scatter([faces[0]['faceLandmarks']['eyeRightBottom']['x']],[faces[0]['faceLandmarks']['eyeRightBottom']['y']])
plt.scatter([faces[0]['faceLandmarks']['eyeLeftOuter']['x']],[faces[0]['faceLandmarks']['eyeLeftOuter']['y']])
plt.scatter([faces[0]['faceLandmarks']['eyeRightOuter']['x']],[faces[0]['faceLandmarks']['eyeRightOuter']['y']])
plt.scatter([faces[0]['faceLandmarks']['eyeLeftInner']['x']],[faces[0]['faceLandmarks']['eyeLeftInner']['y']])
#code for drawing lines
# import numpy as np
# import matplotlib.pyplot as plt

# x, y = np.random.random(size=(2,10))

# for i in range(0, len(x), 2):
#     plt.plot(x[i:i+2], y[i:i+2], 'ro-')

# plt.show()
#end for the code for drawing line
for face in faces:
    fr = face["faceRectangle"]
    fa = face["faceAttributes"]
    origin = (fr["left"], fr["top"])
    p = patches.Rectangle(origin, fr["width"], fr["height"], fill=False, linewidth=2, color='b')
    ax.axes.add_patch(p)
    plt.text(origin[0], origin[1], "%s, %d"%(fa["gender"].capitalize(), fa["age"]), fontsize=20, weight="bold", va="bottom")
_ = plt.axis("off")

image.save('/home/saurabh/Desktop/Hackathon/diagnosis_result/diagnosis_1.png')
```
